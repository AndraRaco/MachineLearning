```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
```

```
def compute_y(x, W, bias):
    # dreapta de decizie
    # [x, y] * [W[0], W[1]] + b = 0
    return (-x * W[0] - bias) / (W[1] + 1e-10)
```

```
def plot_decision_boundary(X, y , W, b, current_x, current_y):
    x1 = -2
    y1 = compute_y(x1, W, b)
    x2 = 2
    y2 = compute_y(x2, W, b)
    # sterge continutul ferestrei
    plt.clf()
    # ploteaza multimea de antrenare
    color = 'r'
    if(current_y == -1):
        color = 'b'
    plt.ylim((-1, 2))
    plt.xlim((-1, 2))
    plt.plot(X[y == -1, 0], X[y == -1, 1], 'b+')
    plt.plot(X[y == 1, 0], X[y == 1, 1], 'r+')
    # ploteaza exemplul curent
    plt.plot(current_x[0], current_x[1], color+'s')
    # afisarea dreptei de decizie
    plt.plot([x1, x2] ,[y1, y2], 'black')
    plt.show(block=False)
    plt.pause(0.3)
```

```
# Date de antrenare

X = np.array([ [0, 0], [0, 1], [1, 0], [1, 1] ])
y = np.array([-1, 1, 1, 1])

# X = np.array([ [0, 0], [0, 1], [1, 0], [1, 1] ])
# y = np.array([-1, 1, 1, 1])

print(X.shape, y.shape)
```

```
# Initializarea ponderilor

W = np.random.random((2))
b = np.random.random((1))

# W = np.random.random((2))
# b = np.random.random((1))

print(W, b)
```

```
plot_decision_boundary(X, y, W, b, X[1], y[1])
```

```
# Antrenare

EPOCHS = 50
EXAMPLES = X.shape[0]
LR = 0.1
```

```
print(X[0].T.shape)
print(W.shape)
```

```
for e in range(EPOCHS):
    X, y = shuffle(X, y)
    for t in range(EXAMPLES):
        y_hat = np.sum(X[t] * W) + b # i
        # print(y_hat)
        loss = (y_hat - y[t]) * (y_hat - y[t]) / 2 # ii
        # print(loss)
        W = W - LR * (y_hat - y[t]) * X[t] # iii
        b = b - LR * (y_hat - y[t]) # iv
        # print(W, b)
        plot_decision_boundary(X, y, W, b, X[t], y[t]) # v
        plt.show()
```

```
# Predictii

y_hat = np.sign(np.dot(X, W) + b)
print(y_hat)
```
